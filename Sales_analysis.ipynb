```python
#GENDER
```

```python
ax = sns.countplot(x = 'Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Gender', y= 'Amount', data = sales_gen)
```

```python
#Hence we can conclude that females have purchased more as compared to males
##Also the total purchasing amount of females is more than that of male
```

```python
#Age
```

```python
ax = sns.countplot(data = df , x= "Age Group")
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
 ax = sns.countplot(data = df , x= "Age Group", hue= 'Gender')
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Age Group', y= 'Amount', data = sales_age)
```

```python
#We can conclude that most of the buyers are of the age group 26-35(females being more as compared to males in the respective age group) and the least buyers belonging to age group 0-17
```

```python
#State
```

```python
ax = sns.countplot(data=df, x="State")
sns.set(rc={'figure.figsize': (25, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x= 'State', y= 'Amount', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x= 'State', y= 'Orders', data = sales_state)
```

```python
#We can conclude that most of the orders are from uttar paradesh, maharashtra and karnataka.
#Key insight is that even though Kerela is ranked 8th in the total amount it is not even in the top 10 in the total number of orders  
```

```python
#Marital status
```

```python
ax = sns.countplot(data = df , x= "Marital_Status")
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(x= 'Marital_Status', y= 'Amount', data = sales_state, hue ='Gender')
```

```python
#We can conclude that married women have done the most shopping and the purchasing amount is also more for most married women
```

```python
##Occupation
```

```python
ax = sns.countplot(data = df , x= "Occupation")
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x= 'Occupation', y= 'Amount', data = sales_state)
```

```python
#We can conclude that most of the buyers work in the IT sector and least in the Agriculture sector 
```

```python
#Product Category
```

```python
ax = sns.countplot(data = df , x= "Product_Category")
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
   ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x= 'Product_Category', y= 'Amount', data = sales_state)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Product_Category', y= 'Amount', data = sales_state)
```

```python
#We can conclude that the most sales are from the food, Clothing&Apparel and Electronics&Gadget Department
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Product_ID', y= 'Orders', data = sales_state)
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Product_ID', y= 'Orders', data = sales_state)
```

```python
#Overall Conclusion:
# Married women aged 26-35 from Uttar Pradesh, Maharashtra and Karnataka working in IT, healthcare and Aviation sectors are more likely to buy 
# products from Food, Clothing and Electronics Category
```
